## **Электромобили в Вашингтоне** 
Я проанализирую данные о зарегистрированных электромобилях в штате Вашингтон с использованием датасета! Ура!

```python
import pandas as tachki

def read_data(file1):
    data = tachki.read_csv(file1)
    return data

def top_brands(data):
    brand = data['Make'].value_counts()
    return brand

def top_models(data):
    model = data['Model'].value_counts()
    return model

def electric_year(data):
    e_types = data.groupby(['Model Year', 'Make']).size()
    return e_types

# кол-во электромобилей, выпущенных в определенный год:
def electric_year_count(data):
    ev_count = data['Model Year'].value_counts().sort_index()
    return ev_count

file1 = 'Electric_Vehicle_Population_Data.csv'
data = read_data(file1)

# самые популярные марки электромобилей
top_brands_result = top_brands(data)
print("Самые популярные марки электромобилей:\n", top_brands_result)

# самые популярные модели электромобилей
top_models_result = top_models(data)
print("\nСамые популярные модели электромобилей:\n", top_models_result)

# типы мобилей по годам
electric_year_result = electric_year(data)
print("\nТипы электромобилей по годам:\n", electric_year_result)

# количество электромобилей по годам
electric_year_count_result = electric_year_count(data)
print("\nКоличество электромобилей по годам:\n", electric_year_count_result)
```

Захотел использовать matplotlib для полноты картины исследования, поэтому убрал дублирование функций с одинаковыми именами electric_year, также создал новую функцию analysis, в которой объединены все анализы, включая создание немного уродливого графика:

```python
import pandas as tachki
import matplotlib.pyplot as graf

def read_data(file1):
    data = tachki.read_csv(file1)
    return data

def top_brands(data):
    brand = data['Make'].value_counts()
    return brand

def top_models(data):
    model = data['Model'].value_counts()
    return model

def analysis(data):
    
    # самые популярные марки электромобилей
    top_brands_result = top_brands(data)
    print("Самые популярные марки электромобилей:\n", top_brands_result)

    # самые популярные модели электромобилей
    top_models_result = top_models(data)
    print("\nСамые популярные модели электромобилей:\n", top_models_result)

    # типы мобилей по годам
    e_types = data.groupby(['Model Year', 'Make']).size()
    e_types.plot(kind='bar', stacked=True)
    graf.title('Типы электромобилей за разные года')
    graf.xlabel('Model Year')
    graf.ylabel('Количество')
    graf.show()

    # количество электромобилей по годам
    ev_count = data['Model Year'].value_counts().sort_index()
    print("\nКоличество электромобилей по годам:\n", ev_count)

file1 = 'Electric_Vehicle_Population_Data.csv'
data = read_data(file1)

analysis(data)
```
